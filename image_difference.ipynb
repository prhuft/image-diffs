```python
from PIL import Image
from numpy import asarray,copy,nan
from IPython.core.display import Image as Im
```

```python
def get_imdiff(fname, fout):
    """
    fname is an image made up of two same-dimensional images pasted side-by-side
    
    this will split the image down the y axis and take the difference of the sides
    and save the image to
    """
    
    im = Image.open(fname).convert('L')

    imarr = asarray(im)
    ncols = imarr.shape[1]

    # make sure even number of columns
    if ncols % 2:
        imarr = imarr[:,:-1] # shave off the last col

    ncols = imarr.shape[1]
    midpt = int(ncols/2)

    larr = imarr[:,:midpt]
    rarr = imarr[:,midpt:]
    
    # take the difference, save and display the image
    darr = rarr - larr
    diff = Image.fromarray(darr).save(fout)

    return darr
```

```python
darr = get_imdiff('platter.jpg', 'platter_diff.png')
```
